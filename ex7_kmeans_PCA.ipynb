### 1.获取每个样本所属的类别，分类

```python
def find_centroids(X, centros):
    idx = []
    
    for i in range(len(X)):
        #求范数
        dist = np.linalg.norm((X[i]-centros), axis=1)
        
        if i ==0:
            print((X[i] - centros), X[i]) #4test
        #返回最小范数的下标   
        id_i = np.argmin(dist)
        idx.append(id_i)
        
    print("第一组数据的分组是：",idx[0:1]) #4test
    
    return np.array(idx)
```

```python
#指定三个大概的中心， k=3
#算出每个样本点到三个中心的距离（即范数）
#找到最小的范数下标，完成了初步分类
# centros = np.array([[0,1], [8,1], [4,5]])
centros = np.array([[0,0], [0,5], [8,5]])
#分组标签
idx = find_centroids(X, centros)
```

### 2.训练聚类中心点

```python
def compute_centros(X,idx,k):
    
    centros = []
    
    for i in range(k):
        #bool下标取值法
        centros_i = np.mean(X[idx == i], axis=0)
        centros.append(centros_i)
        
    return np.array(centros)
```

```python
compute_centros(X, idx, 3)
```

### 3.运行kmeans，重复执行1和2步

```python
#分类点的变化轨迹
def run_kmeans(x, centros, iters):
    k = len(centros)
    centros_all = [] #保存中心点的变化轨迹
    centros_all.append(centros)
    
    for i in range(iters):
        idx = find_centroids(x, centros)
        centros_i = compute_centros(x, idx, k)
        centros_all.append(centros_i)
    
    return idx, np.array(centros_all)
```

```python
idx, centros_all = run_kmeans(X, centros, iters=100)
```

### 绘制数据集和聚类中心的移动轨迹

```python
def plot_data(x, centros_all, idx):
    plt.figure()
    plt.scatter(x[:,0], x[:,1], c=idx, cmap='rainbow')
    plt.plot(centros_all[:,:,0], centros_all[:,:,1], 'kx--')
```

```python
plot_data(X, centros_all, idx)
```

###  观察初始聚类点的位置对聚类效果的影响

```python
def init_centros(x, k):
    #从300个样本数据中，随意选择三个作为初始分类点
    index = np.random.choice(len(x), k)
    
    return x[index]
```

```python
init_centros(X, k=3)
```

```python
for i in range(4):
    idx, centros_all = run_kmeans(X, init_centros(X, k=3), iters=20)
    plot_data(X, centros_all, idx)
```

### 案例：3使用PCA进行二维数据的降维 
数据集：ex7data1.mat

```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
```

```python
data_path = '/home/alex/test/NG_exercise/ex7_kmeans_PCA'
data = sio.loadmat(data_path + '/ex7data1.mat')
data.keys()
```

```python
X = data['X']
X.shape
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

### 对X去均值化 ,即均值变成零

```python
#数据还原时候，记得加上均值

X_demean = X - np.mean(X, axis=0)
plt.scatter(X_demean[:,0], X_demean[:,1])
plt.show()
#均值明显移到了零处
```

```python
np.mean(X, axis=0)
```

```python
(X - np.mean(X, axis=0))[0:5,]
```

### 计算协方差矩阵

```python
#协方差公式CON（X，Y） = E[(X-EX)(Y-EY)]
c = X_demean.T @ X_demean / len(X)
c
```

### 计算特征值，特征向量 

```python
U, S, V = np.linalg.svd(c)
```

```python
S
```

```python
U
```

```python
U1 = U[:,0]
```

### 降维

```python
X_reduction = X_demean@U1
```

```python
X_reduction.shape,U1.shape
```

```python
plt.figure(figsize=(7,7))
plt.scatter(X_demean[:,0],X_demean[:,1])
plt.plot([0,U1[0]],[0,U1[1]],c='r')
plt.plot([0,U[:,1][0]],[0,U[:,1][1]],c='k')

plt.show()
```

###  还原数据

```python
X_restore = X_reduction.reshape(50,1)@U1.reshape(1,2)
            + np.mean(X, axis=0)
```

```python
plt.scatter(X[:,0], X[:,1])
plt.scatter(X_restore[:,0], X_restore[:,1], c='r')
plt.show()
```

### 案例4 ：使用PCA进行图像的降维，k=36
数据集：ex7faces.mat

```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
```

```python
data_path = '/home/alex/test/NG_exercise/ex7_kmeans_PCA'
data = sio.loadmat(data_path + '/ex7faces.mat')
data.keys()
```

```python
X = data['X']
X.shape
```

图片是以数组形式保存的

```python
def plot_100_images(X):
    fig, axs = plt.subplots(ncols=10, nrows=10, 
                            figsize=(10,10)
                           )
    for c in range(10):
        for r in range(10):
            axs[c,r].imshow(X[10*c+r].reshape(32,32).T,
                             cmap='Greys_r'
                            ) #单通道的灰度图
            axs[c,r].set_xticks([])
            axs[c,r].set_yticks([])
    plt.show()
```

```python
plot_100_images(X)
```

```python
means = np.mean(X, axis=0)
X_demean = X - means
C = X_demean.T@X_demean
U, S, V = np.linalg.svd(C) 
```

```python
U.shape, S.shape, V.shape
```

```python
U1 = U[:,:32]
X_reduction = X_demean@U1
X_reduction.shape
```

```python
X_recover = X_reduction@U1.T + means 
```

```python
plot_100_images(X_recover), plot_100_images(X)
```
